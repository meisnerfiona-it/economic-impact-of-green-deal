## Filtered directly from the API

```python
import requests
import pandas as pd
import itertools

url = "https://ec.europa.eu/eurostat/api/dissemination/statistics/1.0/data/prc_hicp_midx"

params = [
    ("lang", "EN"),
    ("geo", "DE"),
    ("geo", "ES"),
    ("geo", "FR"),
    ("geo", "IT"),
    ("coicop", "CP045"),
    ("sinceTimePeriod", "2015-01"),
    ("untilTimePeriod", "2023-12"),
]

r = requests.get(url, params=params, timeout=30)
r.raise_for_status()

data_energy = r.json()

print(data_energy["label"])
print(data_energy["id"])
print(data_energy["size"])
```

```python
for dim in data_energy["id"]:
    print("\nDIMENSION:", dim)
    print(list(data_energy["dimension"][dim]["category"]["index"].keys())[:20])
```

### Make Dataframe

```python
ids = data_energy["id"]
values_obj = data_energy["value"]

# Obtener códigos ordenados según index
dim_codes = {}
for dim in ids:
    cat_index = data_energy["dimension"][dim]["category"]["index"]
    ordered_codes = [k for k, v in sorted(cat_index.items(), key=lambda x: x[1])]
    dim_codes[dim] = ordered_codes

# Crear todas las combinaciones posibles
all_combinations = list(itertools.product(*(dim_codes[dim] for dim in ids)))

# Construir filas
rows = []
for i, combo in enumerate(all_combinations):
    row = dict(zip(ids, combo))
    
    if isinstance(values_obj, dict):
        row["value"] = values_obj.get(str(i))
    else:
        row["value"] = values_obj[i] if i < len(values_obj) else None
        
    rows.append(row)

df_energy = pd.DataFrame(rows)

print(df_energy.head())
print(df_energy.shape)
print(df_energy.columns.tolist())
```

## Rename columns

```python
df_energy = df_energy.rename(columns={
    "geo": "country",
    "time": "date",
    "value": "energy_price_index"
})
```

```python
df_energy = df_energy[["date", "country", "energy_price_index"]].copy()
```

```python
df_energy["date"] = pd.to_datetime(df_energy["date"], errors="coerce")
df_energy["date"] = df_energy["date"].dt.to_period("M").dt.to_timestamp()
```

```python
df_energy["country"] = df_energy["country"].astype(str).str.strip().str.upper()
```

```python
df_energy.head()
```

## Remove duplicates

```python
df_energy = df_energy.dropna(subset=["date", "country", "energy_price_index"])
df_energy = df_energy.drop_duplicates(subset=["date", "country"])
df_energy = df_energy.sort_values(["country", "date"]).reset_index(drop=True)
```

## Save final CSV

```python
print(df_energy.head())
print(df_energy.tail())
print(df_energy.columns.tolist())
print(df_energy["country"].unique())
print(df_energy["date"].min(), df_energy["date"].max())
print(df_energy.isna().sum())
print(df_energy.duplicated(subset=["date", "country"]).sum())
```
